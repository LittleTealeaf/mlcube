```python
import json
import matplotlib.pyplot as plt

average_count = 50

with open('./agent/training_history.json') as file:
    data = json.load(file)

    averages = []

    for i in range(0,len(data),average_count):
        averages.append(sum(data[i:i+average_count]) / average_count)

plt.yscale('log')
plt.plot([i * average_count for i in range(len(averages))],averages)
```
